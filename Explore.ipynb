```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pylab as plt
from matplotlib.pyplot import figure, show

import warnings;
warnings.filterwarnings('ignore');
```

```python
ipl_raw=pd.read_excel("iplallseasons_refined.xlsx")
```

```python
ipl_raw.head()
```

```python
ipl_raw['toss_match_winner'] = np.where(ipl_raw['Toss_winner']==ipl_raw['Winning_team'],'toss_winner','toss_loser')
```

## Match count by venue

```python
ipl_raw.Match_venue.value_counts()
```

```python
ipl_raw['Date'] = pd.to_datetime(ipl_raw.Match_date)
```

```python
ipl_raw['year']=ipl_raw['Date'].dt.year
```

```python
ipl_raw['month'] = ipl_raw['Date'].dt.month
```

```python
ipl_raw.year.unique()
```

```python
ipl_raw.month.unique()
```

```python
def getScore(x):
    x = str(x).split("/")
    return x[0]
```

```python
def getWickets(x):
    x = str(x).split("/")
    if len(x) >1:
        return x[1]
    else:
        return 0
```

```python
ipl_raw['first_batting_score']=ipl_raw.Team1_score.apply(getScore)
```

```python
ipl_raw['first_batting_wickets']=ipl_raw.Team1_score.apply(getWickets)
```

```python
ipl_raw['second_batting_score']=ipl_raw.Team2_score.apply(getScore)
```

```python
ipl_raw['second_batting_wicktes']=ipl_raw.Team2_score.apply(getWickets)
```

```python
ipl_raw.first_batting_score = ipl_raw.first_batting_score.astype(float)
```

```python
ipl_raw.second_batting_score = ipl_raw.second_batting_score.astype(float)
```

```python
ipl_raw.first_batting_wickets = ipl_raw.first_batting_wickets.astype(int)
```

```python
ipl_raw.second_batting_wicktes = ipl_raw.second_batting_wicktes.astype(int)
```
